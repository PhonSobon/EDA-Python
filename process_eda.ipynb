9. Plot different features against one another (scatter), against frequency (histogram)

```python
# Create a histogram for the 'Year' column
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.hist(df['Academic Year'], bins=20, edgecolor='k', alpha=0.7)  # Adjust the number of bins as needed
plt.title('Histogram of Academic Year')
plt.xlabel('Academic Year')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=df, y='Course', color='#9400FF')
plt.title('Count of Students in each Course')
plt.xlabel('Count')
plt.ylabel('Course')
plt.show()
```

```python
print(df.describe())
```

```python
# Group the data by 'Year' and calculate the count of students for each year
yearly_counts = df['Academic Year'].value_counts().sort_index()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-', color='b')  # Adjust the color and style as needed

plt.title('Count of Students Over the Years')
plt.xlabel('Academic Year')
plt.ylabel('Count of Students')
plt.grid(True)

plt.show()
```

```python
from scipy import stats

# Define the value you want to compare with (e.g., the expected mean)
expected_mean = 2020  # You can replace this with your desired value

# Perform a one-sample t-test
t_stat, p_value = stats.ttest_1samp(df["Academic Year"], expected_mean)

print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")
```
